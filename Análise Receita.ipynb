```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
path = '/Volumes/GoogleDrive/Meu Drive/FIP VBC/1. Ativo/Pipeline/Nvoip/2. Analises/Faturamento Dados.xlsx'
```

```python
df = pd.read_excel(path)
df
```

```python
p = df.pivot(columns = 'Natureza', values = 'Valor')
```

```python
p['Identificador'] = df['Identificador']
```

```python
g = p.groupby('Identificador').sum()
```

```python
g.drop(['Outros', 'PabxCloud', 'Pagamento indevido'], axis = 1, inplace = True)
```

```python
g.fillna(0).corr()
```

```python
plt.figure(figsize = (10,10))

sns.heatmap(g.fillna(0).corr(), cmap = 'Blues', annot = True)

plt.show()
```
